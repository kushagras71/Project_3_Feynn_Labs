```python
from sklearn.cluster import KMeans, SpectralClustering
Sum_of_squared_distances = []
K = range(1,15)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(dataset)
    Sum_of_squared_distances.append(km.inertia_)
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
df1 = dataset.loc[:,['thalachh','chol']]

from sklearn.cluster import KMeans
kmeans1 = KMeans(n_clusters=2)
kmeans1.fit(df1)
ls1 = kmeans1.predict(df1)

plt.scatter(dataset['thalachh'], dataset['chol'],c=ls1)
plt.xlabel('thalachh')
plt.ylabel('chol')
plt.show()
```

```python
df2 = dataset.loc[:,['trtbps','chol']]

from sklearn.cluster import KMeans
kmeans1 = KMeans(n_clusters=2)
kmeans1.fit(df2)
ls2 = kmeans1.predict(df1)
plt.scatter(dataset['thalachh'], dataset['chol'],c=ls2)
plt.xlabel('thalachh')
plt.ylabel('chol')
plt.show()
```
